**Importing Libraries**

```python
data['children'].fillna(0,inplace =True)
```

```python
# Dropping columns due to excessive missing values
data.drop(['company','agent'],axis =1,inplace =True)
```

The columns 'agent' and 'company' has many missing values, so they are dropped.

```python
# Checking for duplicates if any
data.duplicated().sum()
```

```python
# Dropping the duplicate values
data.drop_duplicates(inplace =True)
```

```python
# dropping missing values of country column
data.dropna(subset =['country'],inplace =True)
```

```python
# Converting reservation status date column to proper date format
data['reservation_status_date'] = pd.to_datetime(data['reservation_status_date'])
```

**Visualization**

```python
# Plotting bar plot for cancelling rate by hotel type
# Group by hotel type and calculate the cancellation rate
cancellation_rate = data.groupby('hotel')['is_canceled'].mean()
```

```python
# Plot the cancellation rate using Seaborn for a better style
plt.figure(figsize=(8, 5))
sns.barplot(x=cancellation_rate.index, y=cancellation_rate.values)
plt.grid(which='major', axis='y', linestyle='--', linewidth=0.7)  # Dashed gridlines on the y-axis
plt.title('Cancellation Rate by Hotel Type')
plt.xlabel('Hotel Type')
plt.ylabel('Cancellation Rate')
plt.show()
```

```python
# Barplot for canceled vs non canceled bookings
cancellation_counts = data['is_canceled'].value_counts(normalize=True) * 100

# Create a bar plot using Seaborn
plt.figure(figsize=(8, 6))
sns.barplot(x=cancellation_counts.index, y=cancellation_counts.values, palette='coolwarm')

# Set the labels and title
plt.xticks([0, 1], ['Not Canceled', 'Canceled'])
plt.ylabel('Percentage')
plt.xlabel('Booking Status')
plt.title('Percentage of Canceled vs. Non-Canceled Bookings')

# Show the plot
plt.show()
```

**In hotel bookings, lead time refers to the period between when a reservation is made and the guest’s check-in date**

```python
# boxplot for comparing lead_time across customer_type
bookings_per_month = data.groupby('customer_type')['lead_time'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='customer_type', y='lead_time', data=bookings_per_month)

# Set plot title and labels
plt.title('Lead Times Across Different Customer Types', fontsize=16)
plt.xlabel('Customer Type', fontsize=14)
plt.ylabel('Lead Time (days)', fontsize=14)

# Show the plot
plt.show()
```

```python
# Calculate total counts for adults, children, and babies
total_counts = data[['adults', 'children', 'babies']].sum()

# Create a DataFrame for plotting
total_counts_df = total_counts.reset_index()
total_counts_df.columns = ['Category', 'Total Count']  # Rename for clarity

# Plotting
plt.figure(figsize=(8, 5))
sns.barplot(x='Category', y='Total Count', data=total_counts_df, palette='pastel')

# Set plot title and labels
plt.title('Total Number of Adults, Children, and Babies', fontsize=16)
plt.xlabel('Category', fontsize=14)
plt.ylabel('Total Count', fontsize=14)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.dates as mdates

# Group by month/year and calculate the average ADR
monthly_adr = data.resample('M', on='reservation_status_date')['adr'].mean().reset_index()

# Plotting
plt.figure(figsize=(12, 6))
sns.lineplot(x='reservation_status_date', y='adr', data=monthly_adr, marker='o', color='blue')

# Set plot title and labels
plt.title('Average Daily Rate (ADR) Over Time', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Average Daily Rate (ADR)', fontsize=14)

# Format the x-axis to show month names with year
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%B %Y'))

# Show the plot
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```
